```python
def create_stock_data(symbol, granularity, volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength):
    
    try:
        stock_history_object = StockHistory(symbol, granularity = granularity)
        stock_interday_changes = InterdayChanges(stock_history_object)
        stock_interday_metrics = InterdayMetrics(stock_history_object, volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength)


        stock_data = pd.DataFrame()

        stock_data["Date"] = list(stock_history_object.getHistory().keys())[max([volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength]) : -1]
        stock_data["Symbol"] = symbol

        stock_volume_gradient = [stock_interday_metrics.lastVolumeMetric(date)["gradient"] for date in stock_data["Date"]]
        stock_volume_concavity = [stock_interday_metrics.lastVolumeMetric(date)["concavity"] for date in stock_data["Date"]]

        stock_price_gradient = [stock_interday_metrics.lastPriceMetric(date)["gradient"] for date in stock_data["Date"]]
        stock_price_concavity = [stock_interday_metrics.lastPriceMetric(date)["concavity"] for date in stock_data["Date"]]

        stock_pressure_mean = [stock_interday_metrics.lastPressureMetric(date)["mean"] for date in stock_data["Date"]]
        stock_pressure_gradient = [stock_interday_metrics.lastPressureMetric(date)["gradient"] for date in stock_data["Date"]]
        stock_pressure_concavity = [stock_interday_metrics.lastPressureMetric(date)["concavity"] for date in stock_data["Date"]]

        stock_interday_mean = [stock_interday_metrics.interdayMetric(date)["mean"] for date in stock_data["Date"]]
        stock_interday_gradient = [stock_interday_metrics.interdayMetric(date)["gradient"] for date in stock_data["Date"]]
        stock_interday_concavity = [stock_interday_metrics.interdayMetric(date)["concavity"] for date in stock_data["Date"]]
        stock_interday_consecutiveDrops = [stock_interday_metrics.interdayMetric(date)["consecutiveDecreases"] for date in stock_data["Date"]]
        stock_interday_consecutiveJumps = [stock_interday_metrics.interdayMetric(date)["consecutiveIncreases"] for date in stock_data["Date"]]

        stock_data["Volume Gradient"] = stock_volume_gradient
        stock_data["Volume Concavity"] = stock_volume_concavity
        stock_data["Price Gradient"] = stock_price_gradient
        stock_data["Price Concavity"] = stock_price_concavity
        stock_data["Buy Pressure Mean"] = stock_pressure_mean
        stock_data["Buy Pressure Gradient"] = stock_pressure_gradient
        stock_data["Buy Pressure Concavity"] = stock_pressure_concavity
        stock_data["Interday Mean"] = stock_interday_mean
        stock_data["Interday Gradient"] = stock_interday_gradient
        stock_data["Interday Concavity"] = stock_interday_concavity
        stock_data["Consecutive Interday Drops"] = stock_interday_consecutiveDrops
        stock_data["Consecutive Interday Jumps"] = stock_interday_consecutiveJumps

        interday_changes = stock_interday_changes.getInterdayChanges() 

        stock_data["Prev Day Change (%)"] = [interday_changes[date]["prev_day_change"] for date in stock_data["Date"]]
        stock_data["Next Day Change (%)"] = [interday_changes[date]["next_day_change"] for date in stock_data["Date"]]

        return stock_data

    except:
        print("There was an issue with {}".format(symbol))
```

```python
def stock_data(symbols, granularity, volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength):
    
        data = create_stock_data(symbols[0], granularity, volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength)
        
        for symbol in symbols[1:]:
            
            data = data.append(create_stock_data(symbol, granularity, volumeRecordLength, pressureRecordLength, priceRecordLength, interdayRecordLength))
            
        
        return data
```

```python
def remove_nonconsecutive_dates(data):
    
    dates = list(data.query("Symbol == 'AAPL'")["Date"])
    dates_to_keep = []
    
    for i, date in enumerate(dates[:-1]):
        next_date = dates[i + 1]
        if days_between(date, next_date) == 1:
            dates_to_keep.append(date)
    
    return data[data["Date"].isin(dates_to_keep)]
        
```

```python
def create_random_forest(seed = None):
    
    if not seed:
        seed = random.randint(0, 50)
    
    return RandomForestClassifier(class_weight = "balanced", random_state=seed, n_estimators = 50)
    
def create_logistic_regression():
    
    return LogisticRegression(class_weight = "balanced", max_iter = 500, penalty="none")
    

def model_predictions(model, X, threshold = .5):
    
    return model.predict_proba(X)[:, 1] > threshold

    
def cv_model_metrics(model_creator, X, y, threshold):
    
    precisions, adjusted_precisions, accuracies, recalls = [], [], [], []
    
    
    for i in range(10):
        
        model = model_creator()
        rand = random.randint(0, 50)
        X_train, X_test = train_test_split(X, test_size = .1, random_state = rand)
        y_train, y_test = train_test_split(y, test_size = .1, random_state = rand)
        
        model.fit(X_train, y_train.iloc[:, 0])
        
        if type(model) == TwofoldModel:

            model.fit_baseline(X_train, y_train.iloc[:, 1])
        
        y_pred = model_predictions(model, X_test, threshold)
        
        precision = metrics.precision_score(y_test.iloc[:, 0], y_pred, zero_division = 0)
        recalls.append(metrics.recall_score(y_test.iloc[:, 0], y_pred))
        accuracies.append(metrics.accuracy_score(y_test.iloc[:, 0], y_pred))
        precisions.append(precision)
        
        if sum(y_pred) != 0:
            
            adjusted_precisions.append(precisions)
            
    if len(adjusted_precisions) == 0:
        
        return (0, 0), np.mean(accuracies), np.mean(recalls)
    
    else:
        
        return (np.mean(precisions), np.mean(adjusted_precisions)), np.mean(accuracies), np.mean(recalls)
```

```python
class TwofoldModel:
    
    def __init__(self, baseline_threshold):
        self.baseline_threshold = baseline_threshold
        self.primary_model = create_random_forest()
        self.baseline_model = create_random_forest()
        
    def fit(self, X, y):
        self.primary_model.fit(X, y)
        
    def fit_baseline(self, X, y):
        self.baseline_model.fit(X, y)
    
    def predict(self, X):
        
        return self.primary_model.predict(X)*self.baseline_model.predict(X)
    
    def predict_proba(self, X):
        
        raw_predictions = self.primary_model.predict_proba(X)
        auxiliary_predictions = self.baseline_model.predict_proba(X)
        for i in range(X.shape[0]):
            if auxiliary_predictions[i, 1] < self.baseline_threshold:
                raw_predictions[i, 0] = 1
                raw_predictions[i, 1] = 0
        
        return raw_predictions

class ThreefoldModel:
    
    def __init__(self, primary_model, baseline_model_1, basline_model_2, baseline_threshold_1, baseline_threshold_2):
        self.baseline_threshold_1 = baseline_threshold_1
        self.baseline_threshold_2 = baseline_threshold_2
        self.primary_model = primary_model
        self.baseline_model_1 = baseline_model_1
        self.baseline_model_2 = baseline_model_2
        
    def fit(self, X, y):
        self.primary_model.fit(X, y)
        
    def fit_baseline(self, X, y):
        self.baseline_model_1.fit(X, y)
        self.baseline_model_2.fit(X, y)
    
    def predict(self, X):
        
        return self.primary_model.predict(X)*self.baseline_model_1.predict(X)*self.baseline_model_2.predict(X)
    
    def predict_proba(self, X):
        
        raw_predictions = self.primary_model.predict_proba(X)
        auxiliary_predictions_1 = self.baseline_model_1.predict_proba(X)
        auxiliary_predictions_2 = self.baseline_model_2.predict_proba(X)
        for i in range(X.shape[0]):
            if auxiliary_predictions_1[i, 1] < self.baseline_threshold_1 and auxiliary_predictions_2[i, 1] < self.baseline_threshold_2:
                raw_predictions[i, 0] = 1
                raw_predictions[i, 1] = 0
        
        return raw_predictions
```

```python
def load_interday_data(granularity, volumeRecordLenth, pressureRecordLength, priceRecordLength, interdayRecordLength, date = "20201220", symbols = None):
    
    data = import_stock_dataframe(granularity, volumeRecordLenth, pressureRecordLength, priceRecordLength, interdayRecordLength, date)
    
    if symbols:
        
        data = data[data["Symbol"].isin(symbols)]
    
    train, test =  train_test_split(data, test_size = .15, random_state = random.randint(1, 100))
    
    return data, train, test

def find_drops(data, drop_threshold):
    
    data = data.copy()
    data["Is Drop"] = data["Next Day Change (%)"] <= -drop_threshold
    data["Is Decrease"] = data["Next Day Change (%)"] < 0
    
    return data.loc[:, "Volume Gradient": 'Prev Day Change (%)'], data[["Is Drop","Is Decrease"]]

def find_jumps(data, jump_threshold):
    
    data = data.copy()
    data["Is Jump"] = data["Next Day Change (%)"] >= jump_threshold
    data["Is Increase"] = data["Next Day Change (%)"] > 0
    
    return data.loc[:, "Volume Gradient": 'Prev Day Change (%)'], data[["Is Jump","Is Increase"]]
    
```

```python
def cv_evaluate_drop_model(data, model_creator, drop_threshold, model_threshold):
    
    X, y_drops = find_drops(data, drop_threshold)
    
    return cv_model_metrics(model_creator, X, y_drops, model_threshold)
    
```

```python
def test_evaluate_drop_model(full_data, model_creator, drop_threshold, model_threshold):
    

    precisions = []
    accuracies = []
    recalls = []
    
    
    for i in range(35):
        
        train, test =  train_test_split(full_data, test_size = .15, random_state = random.randint(1, 100))
        
        train_X, train_y_drops = find_drops(train, drop_threshold)
        test_X, test_y_drops = find_drops(test, drop_threshold)
        
        model = model_creator()
        model.fit(train_X, train_y_drops.iloc[:, 0])
        
        if type(model) == TwofoldModel:
            model.fit_baseline(train_X, train_y_drops.iloc[:, 1])

        y_pred = model_predictions(model, test_X, model_threshold)

        precisions.append(metrics.precision_score(test_y_drops.iloc[:, 0], y_pred))
        accuracies.append(metrics.accuracy_score(test_y_drops.iloc[:, 0], y_pred))
        recalls.append(metrics.recall_score(test_y_drops.iloc[:, 0], y_pred))
        
    
    return (np.mean(precisions), np.var(precisions)), (np.mean(accuracies), np.var(accuracies)), (np.mean(recalls), np.var(recalls))
    
```

```python
def metric_data_for_thresholds(data, model_creator, model_thresholds, drop_thresholds):
    
    plot_data = pd.DataFrame(columns = ["Model Threshold", "Drop Threshold", "Precision", "Accuracy", "Recall"])
    

    for dt in drop_thresholds:
        
        for mt in model_thresholds:
            
            (precision, adjusted_precision), accuracy, recall = cv_evaluate_drop_model(data, model_creator, dt, mt)
            
            print(dt, mt, precision, accuracy, recall)
            
            plot_data.loc[plot_data.shape[0]] = [mt, dt, precision, accuracy, recall]
    
    return plot_data
```

```python
def metric_data_for_Twofold_model(data, baseline_thresholds, model_thresholds, drop_thresholds):
    
    plot_data = pd.DataFrame(columns = ["Baseline Threshold", "Model Threshold", "Drop Threshold", "Precision", "Adjusted Precision", "Accuracy", "Recall", "Single Model Precision", "Single Model Accuracy", "Single Model Recall", "Twofold Improvement"])
    
    for dt in drop_thresholds:
        for bt in baseline_thresholds:
            for mt in model_thresholds:
                
                model_creator = lambda : TwofoldModel(bt)
                (precision, adjusted_precision), accuracy, recall = cv_evaluate_drop_model(data, model_creator, dt, mt)
                (sm_precision, _), sm_accuracy, sm_recall = cv_evaluate_drop_model(data, lambda : create_random_forest(), dt, mt)
                
                if sm_precision == 0:
                    twofold_improvement = 0
                else:
                    twofold_improvement = 100*(precision - sm_precision)/sm_precision
        
                print("dt: {0}, bt: {1}, mt: {2}, p: {3}, ap: {4}, a: {5}, r: {6}".format(
                    round(dt, 2), round(bt, 2),  round(mt, 2), round(precision, 3), round(adjusted_precision, 3), round(accuracy, 3), round(recall, 3)))

                plot_data.loc[plot_data.shape[0]] = [bt, mt, dt, precision, adjusted_precision, accuracy, recall, sm_precision, sm_accuracy, sm_recall, twofold_improvement]
                
    
    return plot_data

        
```

```python
class DisjointSet:
    
    def __init__(self, elements):
        
        self.element_dct = {}
        self.parent_array = []
        
        for i, element in enumerate(elements):
            self.element_dct[element] = i
            self.parent_array.append(-1)
            
    def connect(self, element1, element2):
        
        assert element1 in self.element_dct and element2 in self.element_dct, "Elements must be in set"
        
        if not self.is_connected(element1, element2):
            
            parent1 = self.find(element1)
            parent2 = self.find(element2)
            
            self.parent_array[parent2] = parent1
        
    def is_connected(self, element1, element2):
        
        return self.find(element1) == self.find(element2)
        
    def find(self, element):
        loc = self.element_dct[element]
        while self.parent_array[loc] >= 0:
            loc = self.parent_array[loc]
        return loc
    
    def list_sets(self):
        
        set_dct = {}
        
        for i, loc in enumerate(self.parent_array):
            if loc < 0:
                set_dct[i] = []
        
        for element in self.element_dct:
            
            parent = self.find(element)
            set_dct[parent].append(element)
        
        return list(set_dct.values())
            
        
```

```python
def find_stock_clusters(singular_value_data, singular_value_count, cluster_count, iter_count):
    
    rand = random.randint(0, 75)
    
    symbol_list_1 = list(singular_value_data.index)[:-1]
    symbol_list_2 = list(singular_value_data.index)[1:]
    
    pair_dct = {}
    for i, symbol1 in enumerate(symbol_list_1):
        for symbol2 in symbol_list_2[i:]:
            pair_dct[symbol1 + "-" + symbol2] = 0
    
    for i in range(iter_count):
        
        clusters = KMeans(n_clusters = cluster_count).fit(singular_value_data.iloc[:, :singular_value_count])
        sv_copy = singular_value_data.copy()
        sv_copy["Cluster"] = clusters.labels_
        for i in range(cluster_count):
            cluster = sv_copy.query("Cluster == {}".format(i))
            
            cluster_symbols_1 = list(cluster.index)[:-1]
            cluster_symbols_2 = list(cluster.index)[1:]
            
            for j, symbol1 in enumerate(cluster_symbols_1):
                for symbol2 in cluster.index[j:]:
                    if symbol1 != symbol2:
                        pair_dct[symbol1 + "-" + symbol2] += 1
    
    sets = DisjointSet(list(singular_value_data.index))
        
    for i, symbol1 in enumerate(symbol_list_1):
        for symbol2 in symbol_list_2[i:]:
            
            if pair_dct[symbol1 + "-" + symbol2]/iter_count >= .90:

                sets.connect(symbol1, symbol2)
                
    
    set_list = sets.list_sets()
    return sorted(set_list, key = lambda x : -len(x))
                    
```
